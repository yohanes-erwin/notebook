```python
import pandas as pd
import numpy as np
```

Indexing dapat dilakukan dengan:
- Value
- List
- Slicing
- Condition
- Function

### Indexing pada Series

```python
obj = pd.Series(np.arange(5),
                index=["a","b","c","d","e"])
```

```python
obj
```

Indexing dengan value

```python
obj["c"]
```

```python
obj[2]
```

Indexing dengan list

```python
obj[["a","c"]]
```

```python
obj[[0,2]]
```

Indexing dengan slicing

```python
obj["a":"c"]
```

```python
obj[0:3]
```

Indexing dengan conditions

```python
obj[obj<2]
```

Indexing dengan function

```python
obj[lambda x : x<2]
```

Mengganti nilai dengan indexing

```python
obj["b":"c"] = 5
obj
```

### Indexing pada DataFrame

```python
data = pd.DataFrame(np.arange(16).reshape(4,4),
                    index=["London","Paris","Berlin","Istanbul"],
                    columns=["one","two","three","four"])
data
```

Indexing dengan value

```python
data["two"]
```

Indexing dengan list

```python
data[["one","three"]]
```

Indexing dengan slicing

```python
data[1:3]
```

Indexing dengan condition

```python
data[data["four"]>5]
```

Mengganti nilai dengan indexing

```python
data[data<5]=0
data
```

### Indexing dengan `iloc` dan `loc`

Fungsi `.iloc()` menggunakan posisi integer.

```python
data.iloc[1]
```

```python
data.iloc[1,[1,2,3]]
```

```python
data.iloc[[1,3],[1,2,3]]
```

```python
data.iloc[1:3]
```

```python
data.iloc[1:]
```

```python
data.iloc[0:3,1:3]
```

Fungsi `.loc()` menggunakan label.

```python
data
```

```python
data.loc["London","two"]
```

```python
data.loc["Paris",["one","two"]]
```

```python
data.loc[:"Paris","four"]
```

```python
data.loc["London":"Berlin","two":"three"]
```

```python
import pandas as pd
```

Membuat DataFrame dari dictionary of list

```python
data = {"name":["Bill","Tom","Tim","John","Alex","Vanessa","Kate"],
        "score":[90,80,85,75,95,60,65],
        "sport":["Wrestling","Football","Skiing","Swimming","Tennis","Karete","Surfing"],
        "sex":["M","M","M","M","F","F","F"]}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

Membuat DataFrame baru dengan urutan kolom berbeda

```python
df = pd.DataFrame(data, columns=["name","sport","sex","score"])
df
```

Mengambil beberapa baris dari atas atau bawah

```python
df.head()
```

Membuat DataFrame baru dengan kolom dari DataFrame sebelumnya dan juga kolom baru

```python
df = pd.DataFrame(data, columns=["name","sport","gender","score","age"])
df
```

Mengganti penamaan index (baris)

```python
df = pd.DataFrame(data, columns=["name","sport","gender","score","age"],
                  index=["one","two","three","four","five","six","seven"])
df
```

Mengisi kolom dengan list

```python
df = pd.DataFrame(data, columns=["name","sport","gender","score","age"],
                  index=["one","two","three","four","five","six","seven"])
values = [18,19,20,18,17,17,18]
df["age"] = values
df
```

Menambah kolom baru berdasarkan kondisi dari kolom sebelumnya

```python
df["pass"] = df.score>=70
df
```

```python
del df["pass"]
df
```
